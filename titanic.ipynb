```python
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd
```

```python
train_data = pd.read_csv('train.csv')
train_data.head(10)
```

```python
train_data.isnull().sum()
```

```python
train_data = train_data.dropna(subset=['Age'])
```

```python
import matplotlib.pyplot as plt

sex_pivot = train_data.pivot_table(index="Sex",values="Survived")
```

```python
display(sex_pivot)
```

```python
selected_rows = ['Sex','Age','Pclass','Survived']
train = train_data[selected_rows]
```

```python
train
```

```python
train['Sex'] = train['Sex'].replace({'male': 0, 'female': 1})
```

```python
display(train)
```

```python
y_train = train['Survived']
X_train = train.drop('Survived',axis=1)
```

```python
from sklearn import svm
clf = svm.SVC()
```

```python
clf.fit(X_train, y_train)
```

```python
test_data = pd.read_csv('test.csv')
```

```python
selected_rows = ['Sex','Age','Pclass']
X_test = test_data[selected_rows]
X_test['Sex'] = X_test['Sex'].replace({'male': 0, 'female': 1})
```

```python
X_test = X_test.dropna(subset=['Age'])
```

```python
predictions = clf.predict(X_test)
```

```python
X_test['Predictions'] = predictions
```
