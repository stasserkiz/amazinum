```python
import pandas as pd
import matplotlib.pyplot as plt
from apyori import apriori
%matplotlib inline
```

```python
df = pd.read_csv('Groceries_dataset.csv')
df.head ()
```

```python
df = df.groupby(['Member_number', 'Date'])['itemDescription'].apply(list)
df
```

```python
df = df.reset_index(drop=True).to_list()
```

```python
rules = apriori(df, min_support=0.002, min_confidence=0.05, min_lift=1, max_length=3)
rules = list(filter(lambda r: len(r.items) > 1, rules))
```

```python
top_rules = sorted(rules, key=lambda r: max(r.ordered_statistics, key=lambda s: s.lift).lift, reverse=True)[:10]
```

```python
def print_rules(rules):
    for rule in rules:
        print ('rule.items=', list(rule.items))
        print ('rule.support=',rule.support)

        for os in rule.ordered_statistics:
            print ('\titems_base=', list(os.items_base))
            print ('\tlifted_item =', list(os.items_add))
            print ('\tlift=', os.lift)
            print ('\tconfidence (i.e. cond prob {} if {})='.format(list(os.items_add), list(os.items_base)), os.confidence)
            print ('\n')
        
print_rules(rules)
```
